```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('dados_jogos.csv')
df = df[df['season']>2010]
df.head()
```

```python
df.dropna(subset=['team_1_mp', 'team_1_fg',
       'team_1_fga', 'team_1_fg_pct', 'team_1_fg3', 'team_1_fg3a',
       'team_1_fg3_pct', 'team_1_ft', 'team_1_fta', 'team_1_ft_pct',
       'team_1_orb', 'team_1_drb', 'team_1_trb', 'team_1_ast', 'team_1_stl',
       'team_1_blk', 'team_1_tov', 'team_1_pf', 'team_1_pts', 'team_1_ts_pct',
       'team_1_efg_pct', 'team_1_fg3a_per_fga_pct', 'team_1_fta_per_fga_pct',
       'team_1_orb_pct', 'team_1_drb_pct', 'team_1_trb_pct', 'team_1_ast_pct',
       'team_1_stl_pct', 'team_1_blk_pct', 'team_1_tov_pct', 'team_1_usg_pct',
       'team_1_off_rtg', 'team_1_def_rtg', 'team_1_ws','team_2_mp',
       'team_2_fg', 'team_2_fga', 'team_2_fg_pct', 'team_2_fg3', 'team_2_fg3a',
       'team_2_fg3_pct', 'team_2_ft', 'team_2_fta', 'team_2_ft_pct',
       'team_2_orb', 'team_2_drb', 'team_2_trb', 'team_2_ast', 'team_2_stl',
       'team_2_blk', 'team_2_tov', 'team_2_pf', 'team_2_pts', 'team_2_ts_pct',
       'team_2_efg_pct', 'team_2_fg3a_per_fga_pct', 'team_2_fta_per_fga_pct',
       'team_2_orb_pct', 'team_2_drb_pct', 'team_2_trb_pct', 'team_2_ast_pct',
       'team_2_stl_pct', 'team_2_blk_pct', 'team_2_tov_pct', 'team_2_usg_pct',
       'team_2_off_rtg', 'team_2_def_rtg', 'team_2_ws'], inplace = True )
```

```python
df['date_game'] = pd.to_datetime(df['date_game'])
df = df.sort_values(['season', 'date_game']).reset_index(drop=True)
```

```python
home_df = df[['team_1', 'season', 'date_game', 'game_result']].copy()
home_df = home_df.rename(columns={'team_1': 'team'})
home_df['game_role'] = 'home'
home_df['result'] = home_df['game_result']

away_df = df[['team_2', 'season', 'date_game', 'game_result']].copy()
away_df = away_df.rename(columns={'team_2': 'team'})
away_df['game_role'] = 'away'
# Visitante ganha se mandante perde
away_df['result'] = away_df['game_result'].apply(lambda x: 'W' if x == 'L' else 'L')

home_df = home_df.sort_values(['team', 'season', 'date_game']).reset_index(drop=True)
away_df = away_df.sort_values(['team', 'season', 'date_game']).reset_index(drop=True)

def compute_streak(df):
    streaks = []
    team_streak = {}
    for team, result in zip(df['team'], df['result']):
        prev = team_streak.get(team, 0)
        if result == 'W':
            if prev > 0:
                new = prev + 1
            else:
                new = 1
        else:
            if prev < 0:
                new = prev - 1
            else:
                new = -1
        streaks.append(new)
        team_streak[team] = new
    return streaks

def compute_momentum(df):
    momentums = []
    team_momentum = {}
    for team, result in zip(df['team'], df['result']):
        prev = team_momentum.get(team, 0)
        new = prev + 1 if result == 'W' else prev - 1
        momentums.append(new)
        team_momentum[team] = new
    return momentums


home_df['wsh'] = compute_streak(home_df)
home_df['gmh'] = compute_momentum(home_df)

away_df['wso'] = compute_streak(away_df)
away_df['gmo'] = compute_momentum(away_df)


# Merge para cada papel do team_1
df = df.merge(
    home_df[['team', 'season', 'date_game', 'wsh', 'gmh']],
    left_on=['team_1', 'season', 'date_game'],
    right_on=['team', 'season', 'date_game'],
    how='left'
).drop(columns=['team'])

df = df.merge(
    away_df[['team', 'season', 'date_game', 'wso', 'gmo']],
    left_on=['team_1', 'season', 'date_game'],
    right_on=['team', 'season', 'date_game'],
    how='left'
).drop(columns=['team'])

df = df.rename(columns={
    'wsh': 'team_1_wsh',
    'gmh': 'team_1_gmh',
    'wso': 'team_1_wso',
    'gmo': 'team_1_gmo'
})

# Agora para o team_2
df = df.merge(
    home_df[['team', 'season', 'date_game', 'wsh', 'gmh']],
    left_on=['team_2', 'season', 'date_game'],
    right_on=['team', 'season', 'date_game'],
    how='left'
).drop(columns=['team'])

df = df.merge(
    away_df[['team', 'season', 'date_game', 'wso', 'gmo']],
    left_on=['team_2', 'season', 'date_game'],
    right_on=['team', 'season', 'date_game'],
    how='left'
).drop(columns=['team'])

df = df.rename(columns={
    'wsh': 'team_2_wsh',
    'gmh': 'team_2_gmh',
    'wso': 'team_2_wso',
    'gmo': 'team_2_gmo'
})
```
