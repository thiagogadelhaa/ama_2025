```python
import pickle
import pandas as pd
```

```python
with open('./dados.pkl', 'rb') as f:
    dados = pickle.load(f)

df_scrapping = pd.DataFrame(dados)
df_scrapping = df_scrapping[df_scrapping['season']>2003]
df_scrapping.head()
```

```python
df_completo = pd.read_csv('./dados_jogos_pt_1.csv')
df_completo = df_completo[df_completo['season']>2003]

df_completo.head()
```

```python
df_scrapping.columns
```

```python
years = list(df_scrapping['season'].unique())
team_1_columns = ['team_1_mp', 'team_1_fg', 'team_1_fga',
       'team_1_fg_pct', 'team_1_fg3', 'team_1_fg3a', 'team_1_fg3_pct',
       'team_1_ft', 'team_1_fta', 'team_1_ft_pct', 'team_1_orb', 'team_1_drb',
       'team_1_trb', 'team_1_ast', 'team_1_stl', 'team_1_blk', 'team_1_tov',
       'team_1_pf', 'team_1_pts', 'team_1_ts_pct', 'team_1_efg_pct',
       'team_1_fg3a_per_fga_pct', 'team_1_fta_per_fga_pct', 'team_1_orb_pct',
       'team_1_drb_pct', 'team_1_trb_pct', 'team_1_ast_pct', 'team_1_stl_pct',
       'team_1_blk_pct', 'team_1_tov_pct', 'team_1_usg_pct', 'team_1_off_rtg',
       'team_1_def_rtg']

team_2_columns = ['team_2_mp', 'team_2_fg', 'team_2_fga',
       'team_2_fg_pct', 'team_2_fg3', 'team_2_fg3a', 'team_2_fg3_pct',
       'team_2_ft', 'team_2_fta', 'team_2_ft_pct', 'team_2_orb', 'team_2_drb',
       'team_2_trb', 'team_2_ast', 'team_2_stl', 'team_2_blk', 'team_2_tov',
       'team_2_pf', 'team_2_pts', 'team_2_ts_pct', 'team_2_efg_pct',
       'team_2_fg3a_per_fga_pct', 'team_2_fta_per_fga_pct', 'team_2_orb_pct',
       'team_2_drb_pct', 'team_2_trb_pct', 'team_2_ast_pct', 'team_2_stl_pct',
       'team_2_blk_pct', 'team_2_tov_pct', 'team_2_usg_pct', 'team_2_off_rtg',
       'team_2_def_rtg'
]

dados_processados = []

for index, row in df_scrapping.iterrows():
    dados_jogo = {}
    
    season = row['season']
    dados_jogo['season'] = season
    dados_jogo['date_game'] = row['date_game']
    team_1 = row['team_1']
    if team_1 == 'CHH' and season<2003:
        continue
    dados_jogo['team_1'] = team_1
    team_2 = row['team_2']
    if team_2 == 'CHH' and season<2003:
        continue
    dados_jogo['team_2'] = team_2
    
    game_index_team_1 = row['g']

    
    query = df_completo[(df_completo['season'] == row['season']) & (df_completo['date_game'] == row['date_game']) &
                (df_completo['team_1'] == team_2) & (df_completo['team_2'] == team_1)]
    

    
    game_index_team_2 = query.iloc[0]['g']
    
    if game_index_team_1 != 1 and game_index_team_2 != 1:
        # Verificando se no jogo anterior jogou fora de casa
        query_team_1 = df_completo[(df_completo['season'] == season) & (df_completo['team_1'] == team_1) & (df_completo['g'] == game_index_team_1-1)]
        
        logic_home_team_1 = query_team_1.iloc[0]['home']

        team_1_colunas = team_1_columns[:] if logic_home_team_1 == 0 else team_2_columns[:]

        if logic_home_team_1 == 0:
            query_team_1_detalhada = df_scrapping[(df_scrapping['season'] == season) & (df_scrapping['team_1'] == team_1) & (df_scrapping['date_game'] == query_team_1.iloc[0]['date_game'])]
            query_team_1_detalhada = query_team_1_detalhada.iloc[0]

            for index, value in enumerate(team_1_colunas):
                dados_jogo[team_1_columns[index]] = query_team_1_detalhada[value]
        else:
            query_team_1_detalhada = df_scrapping[(df_scrapping['season'] == season) & (df_scrapping['team_2'] == team_1) & (df_scrapping['date_game'] == query_team_1.iloc[0]['date_game'])]
            
            for index, value in enumerate(team_1_colunas):
                dados_jogo[team_1_columns[index]] = query_team_1_detalhada[value].iloc[0]

        dados_jogo['team_1_ws'] = query_team_1.iloc[0]['game_streak']
        # Capturando os dados para o team_2
        query_team_2 = df_completo[(df_completo['season'] == season) & (df_completo['team_1'] == team_2) & (df_completo['g'] == game_index_team_2-1)]
        
        logic_home_team_2 = query_team_2.iloc[0]['home']

        team_2_colunas = team_1_columns[:] if logic_home_team_2 == 0 else team_2_columns[:]

        if logic_home_team_2 == 0:
            query_team_2_detalhada = df_scrapping[(df_scrapping['season'] == season) & (df_scrapping['team_1'] == team_2) & (df_scrapping['date_game'] == query_team_2.iloc[0]['date_game'])]

            for index, value in enumerate(team_2_colunas):
                dados_jogo[team_2_columns[index]] = query_team_2_detalhada[value].iloc[0]
        else:
            query_team_2_detalhada = df_scrapping[(df_scrapping['season'] == season) & (df_scrapping['team_2'] == team_2) & (df_scrapping['date_game'] == query_team_2.iloc[0]['date_game'])]
            
            for index, value in enumerate(team_2_colunas):
                dados_jogo[team_2_columns[index]] = query_team_2_detalhada[value].iloc[0]
        
        dados_jogo['team_2_ws'] = query_team_2.iloc[0]['game_streak']
        
        dados_jogo['game_result'] = row['game_result']            
    
    dados_processados.append(dados_jogo)
        
```
